```python
from datetime import datetime
current_time = datetime.now()
formatted_time = current_time.strftime("%d-%m-%y %H:%M:%S")
print("                                       ------------------------------------- ")
print("                                      |Date and time : ", formatted_time,"|")
print("                                       ------------------------------------- ")


class TaskMate:
    def __init__(self):
        self.tasks = []

    def show_tasks(self):
        if not self.tasks:
            print("No tasks added yet!\nDo add a task in your Task-List and start your day!")
        else:
            print("                         Your TaskList : \n")
            for i in range(len(self.tasks)):
                task_str = self.tasks[i]['task']
                status = self.tasks[i]['status']
                print(f"{i + 1}. {task_str} - {'Completed' if status else 'Pending'}")

    def add_task(self):
        task = input("Enter the task to be added in your TaskList : ")
        self.tasks.append({'task': task, 'status': False})
        print("Task added!\nComplete your task in the set time\n")

    def remove_task(self):
        try:
            task_number = int(input("\nEnter task number to remove: "))
            if 0 < task_number <= len(self.tasks):
                confirmation = input(
                    f"Are you sure you want to remove task: '{self.tasks[task_number - 1]['task']}'? (y/n): ")
                if confirmation.lower() == 'y':
                    self.tasks.pop(task_number - 1)
                    print("Task removed!")
                else:
                    print("Task removal canceled.")
            else:
                print("Invalid task number.")
        except ValueError:
            print("Please enter a valid number.")

    def update_task(self):
        try:
            task_number = int(input("\nEnter task number to update: "))
            if 0 < task_number <= len(self.tasks):
                new_task = input(f"Enter new task for: {self.tasks[task_number - 1]['task']}: ")
                self.tasks[task_number - 1]['task'] = new_task
                print("Task updated!")
            else:
                print("Invalid task number.")
        except ValueError:
            print("Please enter a valid number.")

    def mark_complete(self):
        try:
            task_number = int(input("Enter task number to mark as completed: "))
            if task_number > 0 and task_number <= len(self.tasks):
                self.tasks[task_number - 1]['status'] = True
                print(f"Task '{self.tasks[task_number - 1]['task']}' marked as completed!")
            else:
                print("Invalid task number.")
        except ValueError:
            print("Please enter a valid number.")

# Main execution
def main():
    to_do_list = TaskMate()  
    print("                             Welcome to the TaskMake : The Ultimate Task Organizer app ")
    print("                                         Make your task easier to do!")
    while True:  
        print("\n                                              --- TaskList ---")
        print("                                                 1. View Tasks")
        print("                                                 2. Add Task")
        print("                                                3. Remove Task")
        print("                                                4. Update Task")
        print("                                           5. Mark Task as Completed")
        print("                                                   6. Exit\n")
        
        choice = input("Choose an option: ")

        if choice == "1":
            to_do_list.show_tasks()
        elif choice == "2":
            to_do_list.add_task()
        elif choice == "3":
            print("Following are the taskes : \n")
            to_do_list.show_tasks()
            to_do_list.remove_task()
        elif choice == "4":
            print("Following are the taskes : \n")
            to_do_list.show_tasks()
            to_do_list.update_task()
        elif choice == "5":
            to_do_list.mark_complete()
        elif choice == "6":
            print("                                            Have a great day ahead\n                                         Thankyou for using our system!\n!")
            break
        else:
            print("Invalid choice!")

# Run the main function
if __name__ == "__main__":
    main()
```
