```python
to_do_list = []

def show_tasks():
    if not to_do_list:
        print("No tasks added yet!\nDo add a task in your Task-List and start your day!")
    else:
        print("                         Your TaskList : \n")
        for i in range(len(to_do_list)):
            task_str = to_do_list[i]['task']
            status = to_do_list[i]['status']
            print(f"{i + 1}. {task_str} - {'Completed' if status else 'Pending'}")

def add_task():
    task = input("Enter the task to be added in your TaskList : ")
    to_do_list.append({'task': task, 'status': False})
    print("Task added!\nComplete your task in the set time\n")

def remove_task():
    task_number = int(input("\nEnter task number to remove: "))
    if 0 < task_number <= len(to_do_list):
        confirmation = input(f"Are you sure you want to remove task: '{to_do_list[task_number - 1]['task']}'? (y/n): ")
        if confirmation.lower() == 'y':
            to_do_list.pop(task_number - 1)
            print("Task removed!")
        else:
            print("Task removal canceled.")
    else:
        print("Invalid task number.")

def update_task():
    task_number = int(input("\nEnter task number to update: "))
    if 0 < task_number <= len(to_do_list):
        new_task = input(f"Enter new task for: {to_do_list[task_number - 1]['task']}: ")
        to_do_list[task_number - 1]['task'] = new_task
        print("Task updated!")
    else:
        print("Invalid task number.")

def mark_complete():
    task_number = int(input("Enter task number to mark as completed: "))
    if 0 < task_number <= len(to_do_list):
        to_do_list[task_number - 1]['status'] = True
        print(f"Task '{to_do_list[task_number - 1]['task']}' marked as completed!")
    else:
        print("Invalid task number.")


print("                              Welcome to the TaskMake : The Ultimate Task Organizer app ")
print("                                         Make your task easier to do!")
while True:  
    print("\n                                              --- TaskList ---")
    print("                                                 1. View Tasks")
    print("                                                 2. Add Task")
    print("                                                3. Remove Task")
    print("                                                4. Update Task")
    print("                                           5. Mark Task as Completed")
    print("                                                   6. Exit\n")
    choice = input("Choose an option: ")

    if choice == "1":
        show_tasks()
    elif choice == "2":
        add_task()
    elif choice == "3":
        print("Following are the taskes : \n")
        show_tasks()
        remove_task()
    elif choice == "4":
        print("Following are the taskes : \n")
        show_tasks()
        update_task()
    elif choice == "5":
        mark_complete()
    elif choice == "6":
        print("                                            Have a great day ahead\n                                         Thankyou for using our system!\n!")
        break
    else:
        print("Invalid choice!")
```
